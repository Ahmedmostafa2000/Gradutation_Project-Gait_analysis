```python
import pandas as pd
import ast
import numpy as np
import os
```

```python
for j in os.listdir('D:/Career/GradProj/Datasets/Normal_csvFiles'):
    #reading data and changing landmarks type
    df = pd.read_csv('D:/Career/GradProj/Datasets/Normal_csvFiles/'+j)
    df['landmarks'] = df['landmarks'].apply(ast.literal_eval)
    #clearning NaN and reshaping landmarks
    for i in range(len(df.landmarks)):
        try:
            df.landmarks[i] = np.array(list(list(df.landmarks[i].values())[0].values())[23:29]).reshape(-1)
        except:
            df.landmarks[i]=np.NaN
    df.dropna(inplace = True)
    #adding landmarks as columns
    
    df = pd.concat([df, pd.DataFrame(df['landmarks'].to_list())], axis=1)
    df.drop(['landmarks','Unnamed: 0','frame'],axis = 1, inplace = True)
    # try:
    #     df[list(range(0,18,3))]=(df[list(range(0,18,3))]-df[list(range(0,18,3))].min().min())/(df[list(range(0,18,3))].max().max()-df[list(range(0,18,3))].min().min())
    #     df[list(range(1,18,3))]=(df[list(range(1,18,3))]-df[list(range(1,18,3))].min().min())/(df[list(range(1,18,3))].max().max()-df[list(range(1,18,3))].min().min())
    #     df[list(range(2,18,3))]=(df[list(range(2,18,3))]-df[list(range(2,18,3))].min().min())/(df[list(range(2,18,3))].max().max()-df[list(range(2,18,3))].min().min())

    # except:
    #     continue
    
    df.to_csv('D:/Career/GradProj/Datasets/Normal_cleaned/'+j)
```

```python
#concatenation
li = []
for j in os.listdir('D:/Career/GradProj/Datasets/Normal_cleaned'):
    
    df = pd.read_csv('D:/Career/GradProj/Datasets/Normal_cleaned/'+j, index_col=None, header=0)
    
    li.append(df)

dft = pd.concat(li, axis=0, ignore_index=True)
dft.to_csv('D:\Career\GradProj\Datasets\dataset.csv')
```
